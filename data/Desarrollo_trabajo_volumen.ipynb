### 1. ¿Qué Es el Volumen?
El volumen es probablemente el mejor indicador para comprender la popularidad de una criptomoneda.

¿Por qué?

El volumen corresponde al valor de todas las compras y ventas hechas durante un dado período de tiempo, en una o varias exchanges.

En otras palabras:

El volumen será mayor cuanto más transacciones se realicen, lo que muestra el interés general por una determinada criptomoneda.

Imagina que en las últimas 24 horas no hubo un solo intercambio de Bitcoin. En este caso el volumen sería cero. Si un Bitcoin hubiese cambiado de manos dos veces, entonces el volumen sería equivalente al valor de dos Bitcoins.

La popularidad de las principales criptomonedas hace que sea imposible tener volúmenes diarios bajos o nulos. Debido a los miles de intercambios que se realizan todos los días, es natural que el volumen supere varios miles de millones de dólares.

Volúmenes bajos hacen que las criptomoedas sean más vulnerables a grandes variaciones de precio. Esto es especialmente importante a la hora de escoger las criptomonedas en las que invertir.

### 2. Volumen: ¿Un Buen Indicador Técnico?
Como permite medir las transacciones de cada criptomoneda al minuto, el volumen se convierte en un indicador esencial para los traders cuando realizan análisis técnicos, lo que les permite estudiar y anticipar ciertos movimientos de precio.

Por ejemplo:

Si el aumento de precio de una criptomoneda es respaldado por un gran volumen,podría significar que muchas personas están haciendo transacciones o que alguien – una denominada whale – está cambiando grandes cantidades de esa moneda, no sería de extrañar que su valor continúe subiendo;
Por otro lado, si el volumen no sigue el movimiento ascendente, entonces la trayectoria del precio puede ser insostenible, ya que hay poco apoyo para el movimiento.
En resumen:

Además de permitir que se perciba el interés en una determinada criptomoneda, el volumen puede verse como una herramienta fundamental para comprender la aceptación de los inversionistas frente a determinados movimientos de precio.

### 3. Indicador OBV – On Balance Volume
El análisis de los cambios en el volumen para tratar de predecir las tendencias es la base del indicador OBV (On-Balance Volume), desarrollado en 1962 por Joseph Granville, quien creía que el volumen era la gran fuerza que guiaba los mercados.

El OBV es un indicador que suma continuamente el volumen de los días en los que hay variaciones positivas y resta el volumen de los días en donde las variaciones son negativas.

La fórmula para calcularlo es la siguiente:

Si el cierre de hoy es superior al de ayer, entonces: OBV = OBV de ayer + Volumen de hoy
Si el volumen de hoy es inferior al de ayer, entonces: OBV = OBV de ayer– Volumen de hoy
Si el cierre de hoy es igual al de ayer, entonces: OBV = OBV de ayer
En la práctica, hablamos de una serie en la cual son continuamente sumados o restados los volúmenes registrados en los días de análisis.

Si el precio sube y el OBV sube = Confirmación de tendencia alcista
Si el precio baja y el OBV baja= Confirmación de tendencia bajista
Si el precio baja y el OBV sube = Posible reversión de tendencia bajista
Si el precio sube y el OBV baja = Posible reversión de tendencia alcista

### 4. Volume Over the Counter (OTC) – ¿Donde la Magia Ocurre?
Volumen Cambio OTC
¿Será que el volumen presentado en los exchanges nos da una visión real de los montos reales intercambiados diariamente?

La respuesta correcta es no.

Además de las exchanges públicas, hay una forma muy popular de comprar y vender criptomonedas que se llama Over the Counter (OTC).

A diferencia de una transacción facilitada por una exchange, en OTC el intercambio es directo. Dos partes deciden las condiciones bajo las cuales aceptan negociar y completan el proceso de compra y venta sin tener que hacer de forma obligatoria un registro público.

Simplificando…
OTC es una forma de intercambiar criptomonedas en un régimen Peer-to-Peer (P2P), y hay formas muy conocidas de hacerlo.

4.1 LocalBitcoins
LocalBitcoins es en la práctica la democratización de los intercambios de Bitcoin en OTC. A través de una plataforma online, los usuarios pueden encontrar compradores o vendedores que cumplan con sus criterios.

Volumen Localbitcoins p2p
LocalBitcoins ya cuenta con más de 1 millón de usuarios.
«Las personas prefieren comprar y vender Bitcoins OTC, ya que posibilitan transacciones rápidas y con los más diversos métodos de pago. Negociar directamente con otras personas usando el mismo proveedor de pago hace que el dinero se mueva mucho más rápido»

4.2 ¿Cómo es que el volumen OTC influye en el mercado?
El volumen OTC es todo aquél generado por este tipo de intercambios directos.

En el mercado de criptomonedas hay un nombre para los grandes inversionistas: whales o, en español, ballenas.

Son inversionistas que acumulan grandes cantidades de criptomonedas y, por lo tanto, al realizar transacciones provocan grandes alteraciones en el volumen.

Volumen Ballenas Criptomonedas Bitcoin
OTC es una forma de compra o venta popular entre whales.
Por ejemplo, en una exchange pública, una orden de compra o venta de 1.000 Bitcoins llamaría la atención de varios traders y podría provocar alteraciones en el precio. Si estás atento al mercado, podrías cuidar tu posición.

En OTC, a menos que tengas buenas fuentes, la transacción no sería de tu conocimiento.

En otras palabras:
La existencia del mercado OTC es uno de los factores que impiden el acceso a los números totales del volumen de intercambios, no permitiendo que el mercado funcione de la misma manera para todos.

Mientras tanto, los inversionistas que tengan información privilegiada sobre los mercados OTC podrán proteger más fácilmente sus posiciones en los exchages.

4.3 ¿Cuál es el volumen OTC?
El problema para el común de los mortales es precisamente este: si las empresas especializadas en OTC no divulgan esta información, la mayoría de los inversionistas no la conocerán.

Ya que la divulgación de transacciones OTC no es una práctica obligatoria, casi nunca se sabrá con certeza cuál es el verdadero volumen de estas transacciones.

Sin embargo, de acuerdo con un informe del Grupo TABB (acceso pago), el volumen en el mercado OTC de Bitcoin en 2018 podría ser como entre 2 a 3 veces mayor que el de las exchanges.

¡Eso significaría que, en el límite, sólo el 25% del volumen diario de Bitcoin se negociaría en exchanges!

Es decir:

Gran parte del volumen pasaría por el análisis de miles de inversores.

### 5. Volumen Fantasma en Exchanges
Además de los datos de volumen a los que puedes no tener acceso, todavía hay una parte del volumen que no es lo que parece.

Para entender este concepto es esencial tener presente una idea:

Un exchange es tanto más interesante para un inversionista cuanto mayor sea la liquidez que aparenta tener.

Como regla general, los altos volúmenes también suelen indicar altos niveles de liquidez. La liquidez determina con qué facilidad un inversionista puede comprar o vender sus posiciones.

Para explicar esto, vamos a suponer lo siguiente:

Imagina que los vendedores en una exchange no aceptan menos de 10.000$ por un Bitcoin. Los compradores, por su parte, no están dispuestos a pagar más de 5.000$.

Como la diferencia entre los dos precios es grande, se concretarán muy pocas transacciones. Es decir, el volumen será bajo y los inversionistas tendrán dificultades para convertir sus activos en efectivo.

Buenos niveles de liquidez permiten transacciones más rápidas y más eficientes. Cuando el volumen en un exchange es alto, los inversionistas se quedan con la percepción de que sus transacciones se concretarán fácilmente, lo cual es especialmente importante para los inversionistas que hacen day trading.

Por lo tanto, algunas exchanges podrán cuestionarse:

¿Por qué no falsificar los datos del volumen?

Como este no es un mercado regulado, los diferentes actores pueden incurrir en prácticas cuestionables sin mayores consecuencias.

Si bien exchanges como Binance o Coinbase tienen una reputación que defender, otras, más pequeñas, podrán seguir caminos más fáciles para captar la atención de los grandes inversionistas.

5.1 Wash Trading
A la falsificación de los valores del volumen en una exchange se les da el nombre de Wash Trading.

CoinMarketCap agrega datos de cientos de exchanges, pero algunas de ellas pueden estar a proporcionando datos falsos debido a prácticas de Wash Trading, lo que vuelve todavía más difícil tener una idea correcta del volumen en las exchanges.

Otro portal que también agrega información sobre más de 1,000 exchanges es Blockspot.

Como ya hemos visto, los altos volúmenes en el intercambio de una criptomoneda ofrecen a los inversionistas una sensación de mercado que puede ser esencial para tomar decisiones. Si la información a la que tienen acceso no es real, entonces las decisiones que toman fácilmente pueden ser erradas.

En la práctica, una transacción puede fabricarse fácilmente si un inversionista es tanto el vendedor como el comprador de un determinado monto. Si la misma persona está en ambos lados del intercambio, nunca pierde nada, sólo aumenta el volumen en la exchange.

Y, gracias a la tecnología, el proceso es aún más fácil:

«Es fácil para una exchange crear un bot que haga un patrón de trading para que ni siquiera sea necesario hacer estas transacciones a nivel personal.«

Gavin Brown, Profesor de Finanzas en Manchester Metropolitan University, en entrevista a Forbes
Prácticas de este tipo también pueden ocurrir en los mercados tradicionales pero, gracias a la regulación, se convierten en comportamientos más riesgosos, ya que es una práctica prohibida.

5.2 Implicaciones del Wash Trading
Con volúmenes manipulados, las decisiones de inversión se basan en suposiciones falsas, lo que las hace aún más falibles.

Una investigación hecha por Wall Street Journal en octubre de 2018 mostraba que 67% del volumen de intercambios de Bitcoin era pura ilusión.

Ya en 2019, en su informe sobre buenas prácticas, la Blockchain Transparency Institute (BTI) analizó las principales exchanges de acuerdo a CoinMarketCap y llegó a una conclusión impresionante:

¡En 17 de las 25 principales exchanges, Wash Trading llegaba a superar los 99% del volumen presentado!

El Wash Trading es una práctica reprensible que aumenta la incertidumbre en el mercado de las criptomonedas y lo desacredita.

En un momento en que muchos bancos y otras instituciones financieras ya tienen criptomonedas en sus portafolios, quedando más expuestos a este mercado y sus problemas, la falsificación de información puede tener serias consecuencias para la economía global.

5.3 CoinMarketCap da el siguiente paso
Admitiendo el problema del volumen fantasma en exchanges, CoinMarketCap decidió tomar para sí la responsabilidad y presentar una solución.

El portal, que se encarga de agregar información sobre el mercado de las criptomonedas, implementó un indicador de liquidez en las diversas exchanges, como una forma de dar a los inversionistas un indicador menos falible.

Volumen Liquidez
Top 20 exchanges por liquidez.
CoinMarketCap asume en un comunicado que la introducción de este nuevo indicador apunta precisamente a combatir el «problema» de los valores de volumen.

«La métrica sobre la liquidez fue diseñada para, eventualmente, sustituir el volumen como la métrica predefinida al mostrar rankings de pares de criptomonedas y exchanges.»

CoinMarketCap en comunicado
La plataforma también agrega que espera, de esta forma, elevar los standards en el mercado de las criptomonedas, brindando a los inversionistas herramientas para que puedan tomar decisiones más informadas.

En el pasado, ya CoinMarketCap había agregado una sección de ratings con información proporcionada por FCAS (Fundamental Crypto Asset Score).

Conclusión
El volumen es un indicador importante, porque permite saber si hay muchas personas o no comprando una determinada criptomoneda.
El análisis del volumen, tal como ocurre con otros indicadores, puede ser importante al momento de tomar decisiones de inversión más fundamentadas.
El volumen OTC es bastante significativo, pudiendo influir en el mercado, y difícilmente puede ser conocido por la mayoría de los inversionistas.
El volumen fantasma es fabricado por exchanges poco conocidas a través de prácticas de Wash Trading, representando un problema serio que las instituciones ya buscan resolver.
En un mercado que ya es difícil de analizar debido a su gran volatilidad, la posible ausencia de datos de volumen completos, como ocurre debido al OTC Trading, o la introducción de datos forjados, como en el caso del Wash Trading, vuelve todavía más difícil prever movimientos.
Sólo una mayor regulación podrá ayudar a mitigar este tipo de problemas y, así, contribuir a la credibilidad del mercado de las criptomonedas.
