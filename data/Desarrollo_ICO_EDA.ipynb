Criptomonedas
Para las nuevas empresas de alta tecnología y con uso intensivo de capital, el financiamiento es vital para competir y sobrevivir en el mercado. Uno de los pasos fundamentales y, a su vez, más difíciles para los emprendedores, es encontrar y elegir entre diferentes fuentes de capital para financiar su inicio. Actualmente y aprovechando la tecnología blockchain, las startups recientemente comenzaron a financiar sus operaciones con ofertas iniciales de monedas (ICOs). Esto también se conoce a menudo como ventas de tokens.

En las ICO, los empresarios distribuyen activos digitales, como monedas y tokens, a los inversores a cambio de capital. Estos tokens asumen diferentes funciones y utilidades dentro de la red del emisor tan pronto como se lanza el proyecto. Las ofertas iniciales de monedas ICO son una forma novedosa de financiación que ha generado miles de millones de dólares en el ecosistema de la cadena de bloques. Desafiando potencialmente los vehículos de financiación tradicionales, como las inversiones de capital de riesgo o de ángeles comerciales.

¿Qué son las ICOs?
Las ofertas iniciales de monedas ICOs o Token Sales, son una forma de financiación en la que las empresas, principalmente startups, recolectan fondos a través de tokens o criptomonedas que han creado ellas mismas. Funciona de manera similar a una oferta pública inicial tradicional. En el caso de las ICOs, se venden unidades (tokens) de una moneda virtual que aún se encuentran en una fase muy temprana de desarrollo o incluso en el estado de un libro blanco teórico.



Esta venta, para recaudar fondos, suele ser contra Bitcoin (BTC) o Ether (ETH) y no en monedas tradicionales como euros o dólares. La recaudación beneficia a los desarrolladores del proyecto. Además, está destinada a garantizar una viabilidad financiera sostenible para el desarrollo continuo de la moneda virtual que se ofrece. Las ICO son distintas del crowdfunding tradicional en la medida en que la tecnología blockchain transfiere algunas de sus implicaciones y características. Por ejemplo: transparencia, inmutabilidad, descentralización y apertura, a las propiedades de la ICO.

Una oferta inicial de monedas generalmente se anuncia con la divulgación de un documento técnico que describe la venta de tokens, el protocolo de TI subyacente y la cadena de bloques, así como el proyecto y el modelo comercial. También describe la distribución y función del token, los derechos de su titular y su valor. El número de tokens suele ser limitado, con un límite de financiación establecido. La forma más común de ofrecer tokens es a través de una subasta, y los ingresos se destinan a financiar la startup o proyecto.


¿Qué funcionalidad tienen?
Un token ICO es una criptomoneda emitida específicamente para un proyecto en cuestión. Esta tiene entre sus funcionalidades representar digitalmente un derecho o un conjunto de derechos. La forma más sencilla de representación es la del token. Este simple token, llamado de utilidad, no transmite ningún derecho sobre la empresa, el proyecto o el producto. Su incentivo de inversión se debe a la perspectiva real de un aumento de valor en caso de que el proyecto asociado tenga éxito y aumente la demanda de los tokens.

También existe los tokens de uso, que de forma similar a un vale o una licencia, transmiten el acceso o los derechos de uso de un producto o servicio. Un ejemplo de ello es la startup estadounidense Protocol Lab: su Filecoin pretende convertirse en la moneda de una red informática descentralizada en la que los usuarios puedan alquilarse mutuamente el espacio de almacenamiento no utilizado. Sus fichas representan los derechos de uso del espacio de almacenamiento intermediado en su sitio web.

Los ICOs también funcionan como activos o productos. Entre otras cosas, estos tokens pueden encarnar la propiedad de algo. Un ejemplo de ello es la empresa china Tether, cuyo token USDT representa un derecho a uno de los dólares estadounidenses que posee la empresa en cada caso. También pueden funcionar como tokens de trabajo, donde los proveedores no emiten tokens para el pago, sino a cambio de trabajo. Los llamados «acuerdos de asesoría» son aquellos en que se contratan consultores cuyos servicios se recompensan como parte de la venta de tokens.

¿Cómo se puede ganar dinero estando al día de las ICOs?
Antes de invertir en cualquier ICO, debe hacer una investigación exhaustiva sobre las mismas. Ya que actualmente existe una gran lista de ICOs, que por sus cifras proyectadas atraen a una gran cantidad de inversores. Muchas de estas arrojarán algunas cifras muy atractivas, números que seguramente llamarán la atención de cualquier inversor. Nombres como IOTA, Stratis y NEO, que han subido a las filas con ganancias superiores al 100.000%, pueden parecer muy atractivos para el inversor.

Sin embargo, al mirar en retrospectiva, hay una razón mayor por la que estas empresas han encontrado el éxito. Se trata de la visión de la empresa y de los resultados que ofrece a la comunidad. No existe ninguna fórmula que puedas seguir para saber si una ICO será rentable o no. Pero aprender todo lo que puedas sobre el producto, el equipo y la visión de la empresa puede ayudarte a tener tranquilidad con tu inversión.

Para ganar dinero con las ICOs, debes estar al día con la información sobre las mismas y tomar en cuenta estos aspectos para considerar la rentabilidad que pueden proporcionar:


Conocer y hacer un seguimiento al equipo de desarrolladores, las mejores ICOs tienen un gran equipo de desarrolladores detrás de ellas.
Revisar el documento técnico y hoja de ruta definida, esto es importante porque puede ser una manera fácil de determinar si una empresa es legítima o no.
Hacer un seguimiento en redes sociales, revisando las reseñas de la empresa que te interesa.
¿Es seguro participar en ellas?
En vista de que los inversores que se involucran en una ICO, lo hacen en una etapa muy temprana del desarrollo del proyecto, es un negocio de riesgo extremadamente especulativo. Este tipo de negocios está asociado con enormes oportunidades de ganancias, pero también con el riesgo de pérdida total. Si inviertes en una ICO, puedes ganar mucho dinero, ejemplos como NEO, Ethereum o Spectrecoin son una prueba de esto. Pero de ninguna manera cada oferta inicial de monedas se convertirá más tarde en una moneda valiosa.

Por lo tanto, elija su vehículo de inversión con cuidado e invierta solo dinero que pueda recuperar en caso de duda. Este alto riesgo, por un lado, se debe al hecho de que numerosos inversores sin experiencia se sienten atraídos por el fácil acceso a través de Internet y la promesa de dinero rápido. La mayoría de los proyectos consisten en temas técnicos complejos, cuya evaluación requiere una comprensión técnica profunda y una investigación extensa.

Por otro lado, la simplificación técnica del proceso ICO en sí atrae cada vez más a proveedores poco profesionales. Debido a la entrada financiera temprana en el proyecto, es difícil predecir si alguna vez se completará y lanzará, si se desarrollará favorablemente o fracasará más tarde. Además, debido a su internacionalidad, las ICO conllevan el riesgo de una aplicación legal más difícil, ya que los diferentes sistemas e instancias legales a menudo chocan entre sí.

### ¿Cómo podemos sacar rentabilidad a través de las ICO?

Cómo comprar tokens durante una ICO
1. Registrarse para una ICO a través del sitio web del proyecto
Todos los proyectos legítimos que obtienen fondos a través de una ICO tienen un sitio web, donde especifican de qué se trata el proyecto, sus objetivos, la cantidad de dinero que se necesita, cuánto tiempo durará la campaña de financiación, etc. Este sitio web es donde puedes registrarte para la ICO.

Ten cuidado con las campañas de ICOs que no requieren ningún tipo de registro. Por supuesto, todavía hay una posibilidad de que sea una operación completamente legítima, pero últimamente las ICOs más confiables y de alto perfil requieren que los inversores se registren y pasen por un proceso de KYC.

2. Obtén Bitcoin o Ether
Sí, necesitarás tener una de las dos criptomonedas principales para poder participar en una ICO. Consulta nuestras guías dedicadas sobre la compra de Bitcoin y Ether.

Bitcoin, siendo la criptomoneda más dominante, es aceptada prácticamente en cualquier parte del criptomundo. Sin embargo, como Ethereum ofrece una plataforma Blockchain estable y conveniente para que los desarrolladores configuren sus proyectos, se convirtió en una plataforma de elección para las ICO. Entonces, Ether, que es el token nativo de la plataforma Ethereum, se usa ampliamente para comprar tokens durante las ICO.

La cantidad mínima que puedes invertir depende de la ICO en particular. Puede estar entre $10 y $100, o 0,02 ETH a 1 ETH. El monto mínimo de inversión generalmente se indica en el White Paper del proyecto, que se puede encontrar en su sitio web.

3. Mueve tus Bitcoins o Ether a un monedero que controlas
Ya lo hemos dicho innumerables veces: evita mantener tu criptomoneda en un monedero que te proporcione un exchange. Pone en peligro la seguridad de tus fondos, ya que esencialmente no tiene el control de ese monedero. En cambio, mueve tus monedas a un monedero de software que mantiene tus contraseñas en el dispositivo de su elección. Alternativamente, puedes invertir en un monedero de hardware para mayor seguridad. Puedes leer más acerca de los monederos Bitcoin y Ethereum en nuestras guías especializadas.

Ahora, en lo que respecta al monedero que deberás tener para los tokens de la ICO, hay muchas opciones, pero una característica que debes tener en cuenta al elegir uno es que sea compatible con el estándar de token ERC20.

Es el estándar de token más extendido hasta la fecha. Además, es el estándar oficial para la red Ethereum. Por lo tanto, prácticamente cualquier monedero que soporte Ether será compatible con ERC20.

4. Comprar tokens de la ICO
Una vez que estés registrado para una ICO y tengas tus fondos disponibles y listos, todo lo que necesitas hacer es enviar las criptomonedas correspondientes por el monto de tu inversión a la dirección de la campaña.

5. Participa en una ICO enviando tus criptos a su dirección
El objetivo principal de cada campaña de ICO es obtener tu dinero. Por esa razón, intentan hacer que el proceso sea lo más trivial posible. La mayoría de las veces, el sitio web del proyecto le proporcionará pautas detalladas sobre cómo invertir.

Pero, debes ser extremadamente cuidadoso al enviar tus fondos.

En primer lugar, verifica la dirección del sitio web tantas veces como lo consideres necesario. Puede haber sitios web falsos de ICOs listados como anuncios en la parte superior de los resultados de búsqueda de Google. Esos se verán idénticos a los sitios web reales, con uno o dos símbolos de la dirección siendo ligeramente diferentes.

Bitcoin and tentacles

En segundo lugar, verifica la dirección del monedero del proyecto tantas veces como puedas. Habrá estafadores y phishers que publicarán direcciones falsas en línea, por lo que debes tener mucho cuidado de no terminar enviando tu dinero a un lugar equivocado. Si eso sucede, nunca podrás recuperarlo. Lo más recomendable es que verifiquemos los certificados de seguridad la página y nos cercioremos con miembros del equipo de la ICO sobre la veracidad del portal.

Algunas de las ICOs más populares y atractivas tendrán miles y miles de personas enviando Ether al mismo tiempo, lo que puede ocasionar que la red se inutilice, lo que significa que su transacción podría demorar un poco. Siempre puedes verificar el estado de tu transacción aquí simplemente colocando la dirección de tu monedero en la barra de búsqueda.

Finalmente, asegúrate de incluir un poco de Ether adicional en tu transacción. Este Ether adicional irá en "gas", lo que permite transacciones dentro de la red de Ethereum. Básicamente, en este caso particular, el gas sirve como una tarifa de transacción. La cantidad de gas requerida generalmente se presenta en el sitio web del proyecto.

6. Obtén tokens de ICO a tu dirección
Deberías recibir tus tokens recién comprados en la dirección de tu monedero. Si esto no sucede de inmediato, no te preocupes y ten paciencia. Dependiendo de una campaña en particular, sus tokens pueden llegar inmediatamente, pero a veces puede tomar semanas, meses o incluso más. Además, durante una ICO, muchas cosas pueden retrasarse, algunos plazos se pueden mover y así sucesivamente. Por lo tanto, asegúrate de mantenerte actualizado y considerar comunicarte con otros inversores en plataformas y foros dedicados.

To exchange money for crypto

A veces, cuando recibes los tokens, es posible que no puedas cambiarlos inmediatamente en un exchange. Esto depende de las reglas de una ICO particular. Esta información, así como el tiempo que tendrás que esperar antes de recibir los tokens, generalmente se presenta en el sitio web del proyecto o en su White Paper.

7. Una vez que tengas tus tokens de ICO, descubre cómo almacenarlos
De nuevo, siempre asegúrate de que tus fondos estén seguros.

La mayoría de las ICOs más importantes tomarán las medidas necesarias para que tu token sea respaldado por los principales servicios de monederos, especialmente por el servicio de monedero oficial de Ethereum, MyEtherWallet. Incluso si todavía no es compatible con un token en particular, el servicio te permite agregar un token ERC20 personalizado.

Si estás utilizando un monedero de hardware, cualquier token ERC20 puede transferirse al dispositivo y administrarse a través de MyEtherWallet. Además, los fabricantes de monederos de hardware constantemente actualizan sus listas de criptomonedas compatibles. Si el token en el que invertiste recibe ese soporte, todo lo que necesitas hacer es descargar e instalar la última versión del firmware oficial de los fabricantes.

Cómo comprar tokens después de una ICO
Una vez que el período de la ICO haya terminado, y si fue lo suficientemente exitosa, el token eventualmente se incluirá en una lista de criptomonedas. Qué exchange en particular, depende del proyecto. Para saber qué exchange listará el token, puedes seguir los anuncios del proyecto mismo o consultar la sección Altcoin del Foro de Bitcoin y encontrarás un tema sobre el token en particular. Alternativamente, CoinMarketCap enumera la mayoría de las criptomonedas existentes así como los exchanges donde se pueden comerciar.

Cuando un token se lista en un exchange la mayoría de las veces se comerciará contra Bitcoin y Ether. Tenemos guías extensas sobre cómo vender tanto Bitcoin cuanto Ether, que incluyen secciones sobre transacciones en exchanges.

Compré mis primeros tokens, ¿y ahora qué?
Bueno, no hay una respuesta directa a esa pregunta. Es una inversión, así que una vez que adquiriste tus tokens, básicamente tienes dos opciones: utilizar tus tokens en los productos o servicios que se presentan en la ICO o retenerlos y esperar que se revaloricen con el tiempo.

Utilizar los tokens
Los tokens suelen proporcionar a los inversores un acceso futuro al producto o servicio que están ofreciendo. Por ejemplo, si la plataforma ofrece un servicio de apuestas llevadas en tecnología Blockchain, podrías utilizar esos tokens para apostar. También podría pasar que los tokens se empleen para comprar un producto en digital único que será ofrecido por los desarrolladores del proyecto.

Finalmente, si la ICO estaba en manos de una DAO, una Organización Autónoma Descentralizada, que puede establecerse en la plataforma Ethereum, el titular del token también recibe derecho a votar en las propuestas que dan forma a la organización y determinan la dirección que toma.

Comerciar los tokens
Una vez que el token aparezca en una de los exchanges, podrás comerciarlo. Bitcoin y Ether pueden que sean las únicas opciones de negociación durante bastante tiempo después de enlistarse. Otros pares de comercio podrán ser incluidos más adelante, siempre que el token tenga demanda. Pero ten en cuenta que algunos tokens nunca llegaran a ser enlistados.

Hay muchas ICOs exitosas y rentables. Una vez que la compañía pasa con éxito la etapa de ICO, su valor debería comenzar a aumentar, ya que la empresa se vuelve más confiable y está mejor establecida en el mercado. Al mismo tiempo, el valor del token también aumentará, permitiendo a los inversores obtener mayores ganancias. Sin embargo, esto no siempre pasa.

Lo más probable es que necesites tener paciencia con tu inversión por un tiempo. Es posible que tu inversión se convierta en una pérdida total, o no ver ganancias durante mucho tiempo. Es posible que el token tenga valor de casi cero y luego, pueda producirse un pico rápido y masivo, lo que puede significar que el valor del token seguirá aumentando constantemente o podría ser un evento único, después del cual el valor volverá a prácticamente cero.

Esos picos pueden suceder en cuestión de minutos, y necesitas poder atraparlos. Muchos exchanges tienen alertas u opciones de establecer un “límite de pérdida", así que aprovecha estas al máximo.

¿Quiénes pueden participar en una ICO?
No necesitas mucho para invertir en una ICO. Esencialmente, el paquete de inicio incluye acceso a Internet y algunos Bitcoins o Ether. La tecnología Blockchain permite que las inversiones fluyan desde prácticamente cualquier parte del mundo, con tarifas minúsculas y, por lo general, en pocos minutos.

Pero, y este es un "pero" muy importante, asegúrate de que sea legal en tu jurisdicción. Las ICOs son uno de los aspectos de las criptomonedas que las autoridades de todo el mundo se esfuerzan más por regular. Las ICOs están explícitamente prohibidas en algunos países, como China, Corea del Sur y Estados Unidos, mientras que otras han introducido ciertas regulaciones o están a punto de hacerlo. Por ejemplo, en Tailandia las ICO deben pasar por una revisión gubernamental y allí determinarán si el proyecto puede o no salir al mercado a recoger fondos de los inversores.
